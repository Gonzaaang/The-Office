```python
from stop_words import get_stop_words
# stopwords en espa√±ol
palabras_irrelevantes = get_stop_words('english')
```

```python
def grey_color_func(word, font_size, position, orientation, random_state=None,
                    **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(60, 100)
```

```python
from wordcloud import WordCloud, ImageColorGenerator # crea el wordcloud
from PIL import Image # carga la imagen
import matplotlib.pyplot as plt # grafica imagenes
import numpy as np # obtener la forma de la imagen

# graficar nube de palabras con libreria wordcloud
my_mask = np.array(Image.open(r'C:\Users\gonza\Desktop\Gonza\Proyectos\The Office\logof.png')) # obtiene forma del wordcloud

wc = WordCloud(background_color='white',
               mask=my_mask,
               collocations=False,
               width=1024,
               height=,
               contour_width=3, # especifica el ancho del contorno
               contour_color='black', # especifica el color del contorno
               stopwords=palabras_irrelevantes) # agregar stopwords

# leer el archivo txt
with open(r'C:\Users\gonza\Desktop\Gonza\Proyectos\The Office\dialogos.txt', 'r') as txt_file:
    texto = txt_file.read()
# generar nube de palabras
wc.generate(texto.upper())

# extraer colores de la imagen .png usada
image_colors = ImageColorGenerator(my_mask)
# usar el color de imagen original en la nube de palabras (puedes usar colores random en el wordcloud desactivando el c√≥digo de abajo)
wc.recolor(color_func=grey_color_func)


# usar el color de imagen original en la nube de palabras (puedes usar colores random en el wordcloud desactivando el c√≥digo de abajo)


# graficar la nube de palabras
plt.figure(figsize=(15, 8)) # tama√±o de figura
plt.imshow(wc, interpolation='bilinear')
plt.axis('off') # quitar ejes
wc.to_file('wordcloud.png') # guardar wordcloud
plt.show()
```

```python
from wordcloud import WordCloud, ImageColorGenerator#Se utiliza para extraer el color de la imagen
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

text = open(r'C:\Users\gonza\Desktop\Gonza\Proyectos\The Office\dialogos.txt', 'r').read()#Leyendo texto

images=Image.open(r'C:\Users\gonza\Desktop\Gonza\Proyectos\The Office\dundie.jpg')#Reading imagen de fondo
mask_image=np.array(images)#Imagen convertida en matriz
#Equivalente a mask_image = np.array (Image.open ('C: /Users/Violette/Desktop/data/heart.png'))

wordcloud=WordCloud(background_color="white",width=600, max_words=250, 
                    mask=mask_image).generate(text)#Generar diagrama de nube de palabras

#Extraer el color de la m√°scara
image_colors = ImageColorGenerator(mask_image)
# wordcloud.recolor (color_func = image_colors) #Restablecer funci√≥n de color

image=wordcloud.to_image()

# Mostrar imagen
fig,axes = plt.subplots(1,2)
axes[0].imshow(wordcloud)#No establecer color
axes[1].imshow(wordcloud.recolor(color_func=image_colors))#Configure el color de la nube de palabras con el color de la m√°scara
for ax in axes:
    ax.set_axis_off()#Ir al eje de cada imagen
plt.savefig('myfig.png')
plt.show()
```
