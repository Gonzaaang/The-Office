```python
#Importo librerias

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import stylecloud
import stop_words
```

```python
#Importo CSV

df_office = pd.read_csv(r'C:\Users\gonza\Desktop\Gonza\Proyectos\The Office\The_Office_lines.csv', engine = 'python')
df_office = df_office.set_index('id')
df_office = df_office.drop(columns = ['deleted'])
df_office = df_office.rename(columns={'line_text':'Dialogo',
                                   'speaker':'Personaje',
                                     'season':'Temporada',
                                     'episode':'Episodio',
                                     'scene':'Escena'})
df_office
```

```python
# Top de dialogos por personaje

df_office_dialogo = df_office.groupby('Personaje').count().sort_values(by=['Dialogo'],ascending = False)[:10]
top_nombres_personajes = df_office_dialogo.reset_index()['Personaje'].values
df_office_dialogo[['Dialogo']]
```

```python
## Dialogos por personaje por temporada

df_office_personajes = df_office[df_office['Personaje'].isin(top_nombres_personajes)]
df_office_personajes = df_office_personajes.groupby(['Personaje','Temporada']).count().sort_values(by=['Dialogo'],ascending = False).reset_index()
df_office_personajes = df_office_personajes[['Personaje','Temporada','Dialogo']]
```

```python
## Top 10 personajes

import plotly_express as px
import plotly.graph_objects as go
```

```python
## Mapa de calor

df_quien_habla = df_office.sort_values(['id'])
list_quien_habla = list(df_quien_habla['Personaje'].values)
par_personajes = [list_quien_habla[i:i+2] for i in range(0,len(list_quien_habla),1)]
par_personajes.pop(-1)
pairs=[]
for i in par_personajes:
    pairs.append('-'.join(i))

pairs
```

```python
count_elements = dict((i, pairs.count(i)) for i in set(pairs))
```

```python
data_heatmap = sorted(count_elements.items(), key=lambda x:x[1], reverse=True)
data_heatmap = dict(data_heatmap)

x_list = [i.split('-')[0] for i in data_heatmap]
y_list = [i.split('-')[1] for i in data_heatmap]
z_list = [data_heatmap[i] for i in data_heatmap]

x_array = np.array(x_list)
y_array = np.array(y_list)
z_array = np.array(z_list)

df = pd.DataFrame.from_dict(np.array([x_array,y_array,z_array]).T)
df = df[df[0].isin(top_nombres_personajes)]
df = df[df[1].isin(top_nombres_personajes)]
df.columns = ['X_value','Y_value','Z_value']
df['Z_value'] = pd.to_numeric(df['Z_value'])

conditions = [df['X_value']==df['Y_value'],
              df['X_value']!=df['Y_value']]
values = [0, df['Z_value']]

df['Z_value'] = np.select(conditions, values)

####Scale
df['scale'] = np.log(df['Z_value']+1)
pivotted= df.pivot('Y_value','X_value','scale')

pivotted = pivotted[list(top_nombres_personajes[:10])]
pivotted = pivotted.reindex(list(top_nombres_personajes[:10]))
pivotted = pivotted.T #rows speak to columns
pivotted = pivotted.fillna(0)
pivotted
 
#HEATMAP
fig, ax = plt.subplots(1, 1, figsize = (8,6))
ax = sns.heatmap(pivotted, cbar_kws={'ticks':[]})
sns.color_palette("rocket")
ax.set_ylabel('')    
ax.set_xlabel('')
fig.tight_layout()
fig.savefig('heatmap.png')
```

```python

# Analisis de sentimientos
import nltk

from nltk.sentiment.vader import SentimentIntensityAnalyzer


df_sentimiento = df_office[df_office['Personaje'].isin(top_nombres_personajes)]
df_sentimiento = df_sentimiento[['Personaje', 'Dialogo']]

sid = SentimentIntensityAnalyzer()
df_sentimiento.reset_index(inplace=True, drop=True)
df_sentimiento[['neg', 'neu', 'pos', 'compound']] = df_sentimiento['Dialogo'].apply(sid.polarity_scores).apply(pd.Series)
```

```python
#Grafico 

df_sentimiento = df_sentimiento.groupby('Personaje').mean().round(3).sort_values('pos', ascending=True)
df_sentimiento.reset_index(inplace=True)
n = len(df_sentimiento['pos'])
X = np.arange(n)

#bar plot
fig = plt.figure(figsize = (17, 12))
plt.barh(X, df_sentimiento['pos'], facecolor='#9999ff', edgecolor='white')
plt.barh(X, -df_sentimiento['neg'], facecolor='#ff9999', edgecolor='white')

plt.rcParams.update({'font.size':13})
plt.xlim([-.16,.22])
plt.yticks(ticks=X, labels=df_sentimiento['Personaje'], rotation='0')

plt.xlabel('Score Sentimiento')
plt.title('Promedio score de sentimientos de personajes')
plt.legend(('Sentimiento positivo','Sentimiento negativo'))
plt.grid(False)
plt.savefig('sentiment.png', bbox_inches='tight')
plt.show()

```
